*Vi skal her sammenlikne elastisitetene til to lineæare funksjoner. La oss anta at etterspørselen er gitt ved $X_1^D = 500- 5p$ som vi har sett på tidligere, og funnet etterspørselselastisitet $$\epsilon_1^D = -5\frac{p}{x}.$$*      

*Dette skal vi sammenlikne med $X_2^D = 150- p$.*    

* a) Tegn begge etterspørselsfunskjonene i samme figur. Har du noe grunnlag for å si at den ene er "mer/mindre elastisk enn den andre"?*     
     


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# Definerer etterspørselsfunksjonene
def demand_1(p):
    return 500 - (5*p)

def demand_2(p):
    return 150 - p

# Definerer en linspace for priser
p_num = np.linspace(0, 100, 100)

# Plotter etterspørselsfunksjonene i matplotlib
fig, ax = plt.subplots(figsize=(7, 5))

ax.plot(demand_1(p_num), p_num, label='Etterspørsel 1')
ax.plot(demand_2(p_num), p_num, label='Etterspørsel 2')
ax.legend(loc='best',frameon=False);
plt.show()
```

```python
x, p = sp.symbols('x p')
eq_d1 = sp.Eq(demand_1(p), x)
eq_d2 = sp.Eq(demand_2(p), x)

sol_dict = sp.solve ((eq_d1, eq_d2), (p,x))

display(float(sol_dict[p]))
display(float(sol_dict[x]))
```

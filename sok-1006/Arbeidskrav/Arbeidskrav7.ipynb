## Forelesning 11 - Fortjenestemaksimering I - Kort sikt
Vi har antatt at bedriften minimerer sin kostnad ved å produsere en viss mengde produkt. Gitt faste priser på produktet og innsatsfaktorene er dette det samme som å maksimere fortjeneste for denne mengden. Nå spør vi: hvordan bestemmer bedriften hvor mye som skal produseres til enhver produktpris? Vi er med andre ord på sporet av bedriftens *tilbudskurve*. For å kunne produsere bruker bedriften innsatsfaktorer, og vi kan da utlede bedriftens *faktorettespørsel*. 

### 1. Tilpasning i godemarkedt på kort sikt
På kort sikt vil minst én produksjonsfaktor være fast, og dette vil utgjøre en del av bedriftens faste kostnader. Dersom bedriften produserer er produkt ved å kombinere kapital og arbeidskraft, og har en fast mengde med kapital på kort sikt ($K_0$) vil $rK_0$ være en del av bedriftens faste kostnad. Bedriften kan endre produksjon kun ved å endre mengde arbeidskraft som er brukt i produksjonsprosessen.        
Vi kan skrive fortjenesten som en funksjon av produsert mengde $F(x)$:

$$F(x)=px-(C_f+C_v(x)).\tag{1}$$

Vi kan se at fortjeneste er positiv dersom $px>C_f+C_v(x)$, dvs

$$F(x)>0 \Rightarrow p>\frac{C_f+C_v(x)}{x}=\overline{C}(x).\tag{2}$$

Prisen (gjennomsnittsinntekten) må være større enn gjennomsnittskostnaden for positiv fortjeneste. I tillegg til dette er vi ute etter verdien på x som maksmimerer $F(x)$. Førsteordensbetingelsen for et maksimum er

$$F'(x)=p-C'_v(x)=0\Rightarrow p=C'_v(x). \tag{3}$$

Med ord: prisen (her grenseinntekten) må være lik grensekostnaden for et maksimum. Dersom $p>GK$ kan bedriften øke fortjeneste ved å selge en ekstra enhet, og produksjon kan ikke ha vært optimal. Dersom $p<GK$ har bedriften tapt penger på den siste enheten solgt, og kunne ha redusert produksjon for å øke fortjeneste. Da kan ikke denne produksjonen ha vært optimal. Men forsteordensbetingelsen karakteriserer et ekstremt punkt, og vi bør sjekke at dette punktet er et maksimum. Vi gjør dette ved å se på den andre ordensbetingelsen for et maksimum:

$$F''(x)=-C''_v(x)<0\Rightarrow C''_v(x)>0. \tag{4}$$

Dette sier at grensekostnaden må være økende i maksimumspunktet. (*Forklaring i forelesningen!*).




#### 1.1 Eksempel
La oss illustrere med et eksempel (s. 173 i boka): $p=10, C_f=16, C_v=x^2$.
$$F(x)=10x-(16+x^2)$$
Førsteordensbetingelsen blir sa
$$F'(x)=10-2x = 0 \Rightarrow x=5$$
Andreordensbetingelse er oppfylt ettersom $F''(x)=-2 < 0.$ Da har vi funnet maksimal fortjeneste $F(5) = 10(5)-(16+5^2)=9$.


```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def fortj(p, x):
    return p*x-(16+x**2)
```

```python
x = np.linspace(0,10,100)

# bruk axs for å definere aksene

fig, axs = plt.subplots(2,1, sharex=True, figsize=(7,7))

# plott av linjene

axs[0].plot(x,fortj(10, x),label='Fortjeneste (p=10)')
axs[0].plot(x,fortj(11, x), color= 'green', label='Fortjeneste (p=11)')
axs[1].plot(x,2*x, color="red", label='Grensekostnad')
axs[1].hlines(10,0,10, color="black", label='Grenseinntekt (p=10)')
axs[1].hlines(11,0,10, color="green", label='Grenseinntekt (p=11)')

axs[0].vlines(5,0, 9, linestyles='dashed')
axs[0].hlines(9,0, 5, linestyles='dashed')
axs[1].vlines(5,0,20, linestyles='dashed', clip_on=False)
axs[1].vlines(5.5,0,20, color='green', linestyles='dashed', clip_on=False)
axs[0].vlines(5.5,0,14.25, color='green', linestyles='dashed', clip_on=False)




# bruk løkke for å tegne aksene

for i in range(2):

    axs[i].set_ylabel('NOK', loc='top')
    axs[i].set_xlabel('Kvantum (x)', loc='right')
    axs[i].set(xlim=(0,10),ylim=(0,15))
    axs[i].spines['top'].set_color('none')
    axs[i].spines['right'].set_color('none')
    axs[i].tick_params(labelbottom=True)
    axs[i].legend(loc='best')

# gi en passende tittel

axs[0].set_title('Fortjenestefunksjon', loc='left')
axs[1].set_title('Optimalitetsbetingelsen', loc='left')

fig.suptitle('Fortjenestemaksimering på kort sikt');

```

Det vi ser er at Pris = Grensekostnad gir oss en enkel regel for å finne bedriftens *tilbud* på kort sikt (gitt at andreordensbetingelsen er oppfylt, og at fortjenesten er positiv!). ### 2. Fra produktfunksjon til tilbudsfunksjon

For å produsere må bedriften bruke innsatsfaktorer. La oss anta at disse er Arbeidskraft (N) og Kapital (K), og at kapitalmengden er fast på kort sikt $K=K_0$. Produktfunksjonen på kort sikt er dermed
$$x=f(N,K_0)\tag{5}$$

Vi løser (5) for $N$ slik at vi kan skrive $N(x)$ som angir hvor mye arbeidskraft som må ansettes for å produsere $x$ enheter vare (dette er også en funksjon av $K_0$). Da kan vi skrive kostnadsfunksjonen som
$$ C(x) = rK_0 + wN(x)\tag{6}$$

Da har vi kostnadsfunksjonen på samme form som vi brukte i (1), og kan finne bedriftens tilbud!

#### 2.1 Eksempel
En bedrift produserer en vare med produktfunksjon $x = N^{\frac{2}{5}}K^{\frac{1}{4}}$; videre setter vi $K_0=1, w=3, r=5$. Hvor mye bør bedriften produsere for å maksimere fortjeneste på kort sikt?       
Her kan vi ikke sette opp (1) ovenfor ettersom vi har informasjon om produktfunksjonen, men ikke kostnadsfunksjonen. Dette kan vi få:

```python

import sympy as sp
x,N,p,K = sp.symbols('x N p, K', positive=True, real=True) 
```

```python
def prod_ks(N):
    return N**.4

# løs x=f(N)
N_ks = sp.solve(prod_ks(N)-x,N)[0]
N_ks
```

Vi har at $N=x^{\frac{5}{2}}$. Dette kan vi bruke for å finne kostnadsfunksjonen:

```python
Kostnad=5+3*N_ks
Kostnad
```

```python
# tilbudet gis av likning (3) p=GK
GK = sp.diff(Kostnad, x)
GK
```

Tilbudsfunksjonen er $p=7.5x^{\frac{3}{2}}$, gitt at $p>\overline{C}(x)$ (positiv fortjeneste) og $C''(x)>0$ (andre ordensbetingelse).

```python
AOB=sp.diff(GK,x)
AOB
```

```python
# vi gjør klart for å plotte
Kostnad_fn=sp.lambdify(x, Kostnad)

GK_fn=sp.lambdify(x, GK)

display(Kostnad_fn(x))
display(GK_fn(x))
```

```python
def AC(x):
    return Kostnad_fn(x)/x
# finn lavest x som gir positiv fortjeneste, dvs reservasjonsprisen

lavest_x=sp.solve(AC(x)-GK_fn(x),x)[0]
lavest_x
```

```python
x_num=np.linspace(0.01,10,100)
x_1=np.linspace(float(lavest_x),10,100)
fig, ax = plt.subplots()
ax.set_ylabel('Pris', loc='top')
ax.set_xlabel('Kvantum (x)', loc='right')
ax.set(xlim=(0,2))
ax.set(ylim=(0,20))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

ax.plot(x_num, GK_fn(x_num), label = 'GK')
ax.plot(x_1, GK_fn(x_1), color='red', label = 'Tilbud (p=GK)')
ax.plot(x_num, AC(x_num), label = 'Gjennomsnittskostnad')


ax.set_title('Bedriftens tilbud')
ax.legend(loc='best');
```

Tilbudsfunksjonen er den delen av grensekostnadskurven som ligger over gjennomsnittskostnad.

**Oppgave**           
En bedrift produserer en vare med produktfunksjon $x = N^{\frac{2}{5}}K^{\frac{1}{4}}$ (som i Eksempel 2.1). Vis i en figur hvordan tilbudskurven endres når (i) $K$ økes, (ii) $w$ økes og (iii) $r$ økes. Forklar intuisjonen bak dine funn.

```python
# Løsning

def create_basis():
    fig, ax = plt.subplots()
    ax.set_ylabel('Pris', loc='top')
    ax.set_xlabel('Kvantum (x)', loc='right')
    ax.set(xlim=(0,2))
    ax.set(ylim=(0,20))
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.plot(x_num, GK_fn(x_num), label = 'GK')
    ax.plot(x_1, GK_fn(x_1), color='red', label = 'Tilbud (p=GK)')
    ax.plot(x_num, AC(x_num), label = 'Gjennomsnittskostnad')
    return fig, ax

fig, ax = create_basis()

ax.set_title('Bedriftens tilbud - basisfigur')
ax.legend(loc='best', bbox_to_anchor=(1.5, 0.5));
```

```python
import sympy as sp
x,N,p,K,w,r = sp.symbols('x N p K w r', positive=True, real=True)
```

```python
def prod_ks(N,K):
    return N**.4*K**.25
```

```python
# løs x=f(N)
N_ks_2 = sp.solve(prod_ks(N,K)-x,N)[0]
N_ks
```

```python
# Her må man definere w og den andre variablen tror jeg
Kostnad_2=r*K+w*N_ks_2
Kostnad_2
```

```python
# tilbudet gis av likning (3) p=GK
GK_2 = sp.diff(Kostnad_2,(x))
GK_2
```

```python
AOB_2=sp.diff(GK,x)
AOB_2
```

```python
# vi gjør klart for å plotte
Kostnad_fn_2=sp.lambdify((x,w,r,K), Kostnad_2)

GK_fn_2=sp.lambdify((x,w,r,K), GK_2)

display(Kostnad_fn_2(x,w,r,K))
display(GK_fn_2(x,w,r,K))
```

```python
def AC_2(x,w,r,K):
    return Kostnad_fn_2(x,w,r,K)/x
# finn lavest x som gir positiv fortjeneste, dvs reservasjonsprisen

break_even=sp.solve(AC_2(x,w,r,K)-GK_fn_2(x,w,r,K),x)[2]
break_even
```

```python
break_even_fn = sp.lambdify((w,r,K),break_even)

break_even_fn(w,r,K)
```

```python
# utgangspunkt var K=1, w=3, r=5
x_k= np.linspace(float(break_even_fn(3,5,2)),10,100)


fig, ax = create_basis()
# nytt utgangspunkt K=2, w=3, r=5
ax.plot(x_num, GK_fn_2(x_num,3,2,2), label = 'GK(K=2)')
ax.plot(x_k, GK_fn_2(x_k,3,2,2), color='brown', label = 'Tilbud (K=2)')
ax.plot(x_num, AC_2(x_num,3,5,2), label = 'Gjennomsnittskostnad(K=2)')

ax.set_title('Bedriftens tilbud -økning i K')
ax.legend(bbox_to_anchor=(1, 0.6));
```

```python
x_k= np.linspace(float(break_even_fn(3,6,1)),10,100)

fig, ax = create_basis()
# nytt utgangspunkt K=1, w=3, r=5
ax.plot(x_num, GK_fn_2(x_num,3,6,1), label = 'GK(r=6)')
ax.plot(x_k, GK_fn_2(x_k,3,6,1), color='brown', label = 'Tilbud (r=6)')
ax.plot(x_num, AC_2(x_num,3,6,1), label = 'Gjennomsnittskostnad(r=6)')

ax.set_title('Bedriftens tilbud -økning i r')
ax.legend(bbox_to_anchor=(1, 0.6));
```

```python
x_k= np.linspace(float(break_even_fn(4,5,1)),10,100)

fig, ax = create_basis()
# nytt utgangspunkt K=1, w=3, r=5
ax.plot(x_num, GK_fn_2(x_num,4,5,1), label = 'GK(w=4)')
ax.plot(x_k, GK_fn_2(x_k,4,6,1), color='brown', label = 'Tilbud (w=4)')
ax.plot(x_num, AC_2(x_num,4,5,1), label = 'Gjennomsnittskostnad(w=4)')

ax.set_title('Bedriftens tilbud -økning i w')
ax.legend(bbox_to_anchor=(1, 0.6));
```
